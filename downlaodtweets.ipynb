```python
import tweepy
from tweepy import OAuthHandler
import json
import pandas as pd
```

```python
consumer_key = "XXXXX"
consumer_secret = "XXXXX"
access_token = "XXXXX"
access_secret = "XXXXX"
```

```python
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)
api = tweepy.API(auth, wait_on_rate_limit=True, wait_on_rate_limit_notify=True)
```

```python
def lookup_tweets(tweet_IDs, api):
    full_tweets = []
    tweet_count = len(tweet_IDs)
    try:
        for i in range((tweet_count//100) + 1):
        #for i in range(50):
            # Catch the last group if it is less than 100 tweets
            end_loc = min((i + 1) * 100, tweet_count)
            id=tweet_IDs[i * 100:end_loc]
            full_tweets.extend(api.statuses_lookup(id))
        return full_tweets
    except tweepy.TweepError:
        print('Something went wrong, quitting...')
```

```python
df=pd.read_csv('deleted_tweets_public_share.csv')
```

```python
good_tweet_ids = [i for i in df["tweetID"]]#tweet ids to look up 
results = lookup_tweets(good_tweet_ids, api) #apply function
temp = json.dumps([status._json for status in results]) #create JSON
df1 = pd.read_json(temp, orient='records')
df1.shape
```

```python
df1.to_csv('deleted_tweets.csv', index=False)
```
